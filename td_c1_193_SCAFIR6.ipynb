```python
import grrmlib as gl

gl.__version__
```

```python
eqs = gl.read_eq_list("td_c1_193_SCAFIR6_EQ_list.log")
pts = gl.read_pt_list("td_c1_193_SCAFIR6_PT_list.log")
```

```python
rpn = gl.ReactionPathNetwork(eqs=eqs, pts=pts)
rpn = rpn.remove_eq("??")
```

```python
grpn = rpn.cluster(gl.is_identical)
```

```python
writer = gl.GaussianInputWriter()
```

# 0. Benzene complex

```python
eqs0 = (
    eqs
    .filter(lambda eq: 2.25 < eq.get_distance(1, 25) < 2.65)
    .filter(lambda eq: eq.get_distance(13, 25) > 2.60)
    .filter(lambda eq: eq.get_distance(16, 25) > 2.60)
    .filter(lambda eq: eq.get_distance(7, 37) > 1.10)
    .filter(lambda eq: eq.get_distance(7, 35) > 1.10)
    .filter(lambda eq: eq.get_distance(1, 13) > 1.60)
    .filter(lambda eq: eq.get_distance(14, 25) > 1.60)
    .filter(lambda eq: 3.50 < eq.get_distance(35, 42) < 4.50)
    .filter(lambda eq: eq.get_distance(52, 53) < 1.20)
)
```

```python
len(eqs0)
```

```python
eqs0_min = eqs0.smallest("scfenergy")
```

```python
eqs0_min.name
```

```python
writer.write(eqs0_min, f"EQ{eqs0_min.name}.com", overwrite=True)
```
